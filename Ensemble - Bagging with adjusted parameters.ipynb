# Build Model

Import Machine Learning Packages

```python
# ML Pkg
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier

# Metrics
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

Declare Features and Labels

```python
X_train = df_train.drop(["ID X P_Code", "Target"], axis =1)
y_train = df_train["Target"]

X_test = df_test.drop(["ID X P_Code"], axis = 1)
```

```python
x_train,x_validation,y_train,y_validation = train_test_split(X_train,y_train,test_size=0.20,random_state=7)
```

```python
from sklearn import model_selection
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
kfold = model_selection.KFold(n_splits=10, random_state=7)
cart = DecisionTreeClassifier()
num_trees = 100
model = BaggingClassifier(base_estimator=cart, n_estimators=num_trees, random_state=7)
model.fit(x_train,y_train)
```

```python
prediction = model.predict(x_validation)
```

Evaluate Accuracy and Log_loss

```python
from sklearn.metrics import log_loss, accuracy_score
```

```python
score = accuracy_score(y_validation, prediction)
print(f"Log Loss is: {score}.")
```

Making Prediction on Test Data

```python
Test_prediction = model.predict_proba(X_test)[:,-1] #predicting probability.  showing the ID X P_Code column and the prediction alone
```

```python
pd.DataFrame({"ID X P_Code": df_test["ID X P_Code"], "Label": Test_prediction}).to_csv("Twelveth_Submission.csv", index = False)
#Map the expected user_id column to user_id on the test file, and also the 'Test _prediction' to 'CHURN'
```
